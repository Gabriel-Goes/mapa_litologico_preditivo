```python
chain_list = ['CTC', 'KC', 'UC', 'THC', 'MAGR']

grids = {'CTC':(),'KC':(), 'UC':(), 'THC':(), 'MAGR':()}

scores = {'CTC':(),'KC':(), 'UC':(), 'THC':(), 'MAGR':()}


coordinates = (g_1039_SF23_Y_A_III_4.UTME.values, g_1039_SF23_Y_A_III_4.UTMN.values)
```

```python
for i in chain_list:
    print(i)
    
    chain = vd.Chain([
        ('trend', vd.Trend(degree=1)),
        ('reduce', vd.BlockReduce(np.median, spacing=1000)),
        ('spline', vd.Spline())
    ])
    
    chain.fit(coordinates, g_1039_SF23_Y_A_III_4[i])
  
    cv     = vd.BlockKFold(spacing=1000,
                  n_splits=100,
                  shuffle=True)

    scores[i] = vd.cross_val_score(chain,
                             coordinates,
                             g_1039_SF23_Y_A_III_4[i],
                             cv=cv)

    grid = chain.grid(spacing=200, data_names=[i],pixel_register=True)
    grids[i] = vd.distance_mask(coordinates, maxdist=1000, grid= grid)
```

```python
plt.figure()
plt.hist(scores['CTC'],bins='auto')
```
