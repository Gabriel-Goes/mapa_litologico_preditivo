```python
from src import set_gdb
from tkinter import Tk, Label, Button, ttk
import geopandas as gpd
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
def import_malha_cartog(escala='50k',ID=['SF23_YA','SF23_YB',
                                         'SF23_VC','SF23_VD']):
    mc = gpd.read_file(set_gdb('geodatabase.gpkg'),driver='GPKG',layer='mc_'+escala)
    mc_slct = gpd.GeoDataFrame()
    print(f' - Malha Cartografica : {mc.shape}, Seleção:  {mc_slct.shape}')
    if ID:
        for id in tqdm(ID):
            print(f' - ID: {id}')
            mc_slct.append(mc[mc['id_folha'].str.contains(id)])
            print(mc_slct)
    
        return mc_slct
        
    else:
        print('mc')
        return mc
```

```python
#-----------------------------Main Frame---------------------------- 
janela=Tk()
titulo=Label(text='Seleção de Malha Cartográfica',font=('Arial',15))
titulo.pack()
# ------------------------------------------------------------------
escalas = ['25k','50k','100k','250k','1kk']

# -------------------------Seletor----------------------------------
frame_escala = ttk.Frame()

lbl_escala = ttk.Label(frame_escala,text='Escalas Disponíveis')
cmb_escala = ttk.Combobox(frame_escala,values = escalas)
lbl_escala.grid(row=0,column=0, padx=15)
cmb_escala.grid(row=0,column=1)
cmb_escala.set('1kk')
frame_escala.pack()


# ------------------------------------------------------------------
janela.mainloop()
```

```python
# ----------------------------------- MAINFRAME -----------------------------------------------
janela = Tk()
titulo = Label(
    text='Preditor de Litologias',
    font=('Arial',15)
)
titulo.pack()
# ---------------------------------------------------------------------------------------------
escalas = [
    '25k',
    '50k',
    '100k',
    '250k',
    '1kk'
]
# ----------------------------------- FUNCTIONS -----------------------------------------------
def get_folhas():
    escala = combo_escala.get()
    mc = src.import_malha_cartog(escala)
    list_ids = list(mc['id_folha'])
    print(f' - Escala selecionada: {escala}')
    print(f' - Número de folhas: {len(list_ids)}')
    return list_ids
def build_dic():
    escala = combo_escala.get()
    quadricula = src.Build_mc(escala=escala,)
    print(quadricula.keys())

def Update_dic():
    quadricula = src.Upload_mc()
    print(quadricula.keys())
# ----------------------------------------------------------------------------------------------
#   
# ------------------------------------- Botao Escolha escala -----------------------------------
frame_escala = ttk.Frame()

botao_set_scale = Button(frame_escala,text='Escolha',command=get_folhas)
botao_set_scale.grid(row=0,column=2, padx=15)
label_escala = ttk.Label(frame_escala,text='Escalas Disponíveis')
combo_escala = ttk.Combobox(frame_escala,values = escalas)
combo_escala.set('1kk')
label_escala.grid(row=0,column=0, padx=15)
combo_escala.grid(row=0,column=1)
frame_escala.pack()
# Lista de Folhas Disponíveis
frame_id = ttk.Frame()
label_id = ttk.Label(frame_id, text = 'IDs Disponíves')
combo_id = ttk.Combobox(frame_id,values = get_folhas())
label_id.grid(row=0,column=0,padx=15)
combo_id.grid(row=0,column=1,padx=15)
frame_id.pack()
# Criar Dicionario
botao_create = Button(text='Create Dictionary',command=build_dic)
botao_create.pack()
# Atualizar Dicionário
botao_update = Button(text='Update Dictionary',command=Update_dic)
botao_update.pack()





janela.mainloop()
```
