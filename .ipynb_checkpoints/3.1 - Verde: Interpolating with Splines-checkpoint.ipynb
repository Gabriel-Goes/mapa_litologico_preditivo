```python
fig, axs = (2,3 )

for ax, f in zip(raw_data, dic_titles):
    vmin=data_descrito[f]['1%']
    vmax=data_descrito[f]['99.5%']
    
    #plt.figure(f + " Dados brutos")
    plt(ax)
    
    plt.scatter(data.X, data.Y,
            c       = data[f],
            cmap    = 'rainbow',
            s       = 0.5,vmax=vmax,vmin=vmin)
    plt.colorbar(label= f)
    plt.title(label='Aerogamaespectometria da folha SF23_Y_A_III_4')
    plt.axis('scaled')
```

```python
data['geometry'] = [geometry.Point(x,y) for x, y in zip(data['X'], data['Y'])]
```

```python
data
```

```python
# Block Reduction
reducer_median = vd.BlockReduce(np.mean, spacing= 499)
b_coords, b_CTCOR = reducer_median.filter(coordinates, data.CTCOR)
```

```python
# Ploting the Reduced data
plt.figure(figsize=(9,9))
plt.scatter(b_coords[0],b_coords[1],
            c    = b_CTCOR,
            cmap = 'rainbow',
            s    = 5,
            marker='o')
plt.axis('scaled')
plt.title('Amostras reduzidas para distâncias médias de 500 m')
plt.colorbar(label='Contagens Totais Corrigidas')
```

```python
spline = vd.Spline()
```

```python
spline.fit(b_coords, b_CTCOR)
```

```python
predicted = spline.predict(coordinates)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(coordinates[0],coordinates[1],
            c=predicted,
            cmap='rainbow',
            s=1)
plt.axis('scaled')
plt.title('Valores preditos para os mesmos pontos amostrados')
plt.colorbar(label='CTCOR')
```

```python
residuals = data.CTCOR - predicted
```

```python
scale = vd.maxabs(residuals)

plt.figure(figsize=(8,8))

plt.scatter(coordinates[0],coordinates[1],
            c=residuals,
            cmap='RdBu_r',
            s=2,
            vmin=-scale,vmax=scale)
plt.axis('scaled')
plt.title('Residuals: Dado amostrado - Dado predito')
plt.colorbar(label='CTCOR')
```

```python
region = vd.get_region(coordinates)
grid_coords = vd.grid_coordinates(region, spacing = 100,pixel_register=True)
```

```python
chain = vd.Chain([
    ('trend',  vd.Trend(degree=1)),
    ('reduce', vd.BlockReduce(np.median, spacing=500)),
    ('spline', vd.Spline())
])
```

```python
data.head()
```

```python
chain.fit(coordinates, data.CTCOR)

grid = chain.grid(spacing=100, data_names=['CTCOR'])
grid_mask = vd.distance_mask(coordinates, maxdist=500, grid=grid)
```

```python
grid_mask
```

```python
grid_mask.CTCOR.plot(figsize=(8,8), cmap='hsv')
plt.axis('scaled')
```

```python
train, test = vd.train_test_split(coordinates, data.CTCOR, test_size=0.1, spacing=500)
```

```python
plt.figure()
plt.plot(train[0][0], train[0][1], '.b', markersize=3)
plt.plot(test[0][0], test[0][1], '.r', markersize=3)
plt.axis('scaled')
```

```python
chain.fit(*train)
```

```python
chain.score(*test)
```

```python
cv     = vd.BlockKFold(spacing=100,
                  n_splits=15,
                  shuffle=True)
```

```python
scores = vd.cross_val_score(chain,
                            coordinates,
                            data.CTCOR,
                            cv=cv)
```

```python
scores
```

```python
plt.figure()
plt.hist(scores, bins ='auto')
```
